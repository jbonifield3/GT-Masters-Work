```python
#Loading Python Kernel.....

#The passenger arrives and is put in an Arrival Queue
#The passenger is then assigned to a Scanner queue

import random
import simpy

ArrivalRate = 1/10 #5 arrivals per minute
QueueRate = 2/1 #1 process per 2 minutes
NumQueues = 20 #Number of Arrival Queues
NumScanners = 15 #Number of Scanners
ScannerLB = 0.5 #Scanner Lower Bound time to process
ScannerUB = 1.0 #Scanner Upper Bound time to process


Time = 10 #Time to run the simulation, one day
RandomSeed = 5311994

class Airport(object):

    def __init__(self, env, num_machines, num_scanners):
        self.env = env
        self.checker = simpy.Resource(env, num_machines)
        self.scanner = simpy.Resource(env, num_scanners)    
        
    def check(self, passenger, queue_rate):
        """
        The arrival check process. It takes a passenger processes and checks it.
        """
        yield self.env.timeout(queue_rate)
        
    def scan(self, passenger, scan_lower_bound, scan_upper_bound):        
        """
        The scanning process. It takes a passenger processes after check in and scans it.
        """
        scan_time = random.uniform(ScannerLB,ScannerUB)
        yield self.env.timeout(scan_time)
        
def passenger(env,name,airport):
    '''
    Each passenger in airport env has a Name name, and is sorted into arrival_station.
    After being processed by the arrival_station, passenger is moved to scanning_station
    '''
    print('%s arrives at the Airport at %.2f.' % (name, env.now))
    with airport.checker.request() as request:
        yield request 
        
        print('%s enters the arrival queue at %.2f.' %(name, env.now))
        yield env.process(airport.check(name,QueueRate))
        print('%s leaves the arrival queue at %.2f.' %(name, env.now))
        
    with airport.scanner.request() as request:
        yield request 
        
        print('%s enters the check queue at %.2f.' %(name, env.now))
        yield env.process(airport.scan(name,ScannerLB,ScannerUB))
        print('%s leaves the check queue for the gate %.2f.' %(name, env.now))

def setup(env,arrival_rates,number_of_queues, number_of_scanners):
    """
    Create Airport Object, a number of initial Passengers and keep creating passengers
    approx. every ArrivalRate minutes.
    """
    # Create the Airport
    AtlantaAirport = Airport(env, number_of_queues, number_of_scanners)
    i = 0
    # Create 5 initial cars
    #for i in range(5):
    #    env.process(passenger(env, 'Passenger %d' % i, AtlantaAirport))

    # Create more passengers while the simulation is running
    while True:
        yield env.timeout(random.uniform(arrival_rates - arrival_rates/10, arrival_rates + arrival_rates/10))
        i += 1
        env.process(passenger(env, 'Passenger %d' % i, AtlantaAirport))
        
random.seed(RandomSeed)         
env = simpy.Environment()

env.process(setup(env,ArrivalRate,NumQueues,NumScanners))

# Execute!
env.run(until=Time)
```

From this scenario above, testing for 10 min, we can see that the last passenger to leave the check queue was #30 out of 50 total arrivals. Let's explore what happens if we drop the number of scanners down to 1:

```python
NumScanners = 1

random.seed(RandomSeed)
env = simpy.Environment()
env.process(setup(env,ArrivalRate,NumQueues,NumScanners))
env.run(until=Time)
```

We can see that under these circumstances, only 11 passengers make it through the gates out of the 51. Let's bump the arrivals up to 50, with more scanners and queues:

```python
ArrivalRate = 1/50 #50 arrivals per minute
NumScanners = 15
NumQueues = 20 #Number of Arrival Queues

random.seed(RandomSeed)
env = simpy.Environment()
env.process(setup(env,ArrivalRate,NumQueues,NumScanners))
env.run(until=Time)
```

We can see that this is a pretty sufficient amount of equipment - and based on this model we would probably recommend much more equipment than 15 scanners and 20 lines for a large airport like ATL
